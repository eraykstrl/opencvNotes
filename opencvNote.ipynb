cv2.normalize(roi_hist,roi_hist,0,255,cv2.NORM_MINMAX)
## [hsv_roi]: Histogramın hesaplanacağı görüntü. Burada, hsv_roi adlı bir görüntü kullanılıyor. hsv_roi, renk uzayının HSV (Renk Tonu, Doyma ve Parlaklık) kanalında bulunan bir bölgenin veya nesnenin görüntüsünü içerir.
## [0]: Histogramın hesaplanacağı kanalın indeksi. Renk uzaylarının her biri farklı kanallara sahiptir. Örneğin, HSV renk uzayında renk tonu kanalı 0 indekslidir. Bu, histogramın renk tonu kanalına göre hesaplanacağını belirtir.

## None: Maske. Histogramı hesaplarken bir maske kullanılabilir. Maske, histogramın belirli bir bölgede veya özelliğe sınırlanmasını sağlar. Bu durumda, herhangi bir maske kullanılmıyor (None olarak belirtiliyor), bu nedenle tüm görüntüdeki renk tonu dağılımı hesaplanacak.

## [180]: Histogramın histogramın tarama aralığının (bins) sayısı. Histogram, renk tonu gibi bir değerin belirli aralıklarla taramasıdır. Burada, renk tonu 0 ila 180 aralığında olduğu için 0'dan 180'e kadar olan 180 tarama aralığı kullanılıyor.

## [0, 180]: Histogramın tarama aralığının minimum ve maksimum değerleri. Renk tonu kanalı genellikle 0 ila 180 derece aralığında olduğu için bu aralık belirtiliyor.

    cv2.normalize neden kullanılır 

## cv2.normalize() işlevi, bir veri kümesini belirli bir aralığa yeniden ölçeklemek için kullanılır. İşte bu işlevin argümanlarının anlamı:

## roi_hist: Normalleştirilecek veri kümesi. Bu durumda, daha önce hesaplanmış olan histogramı (roi_hist) yeniden ölçeklemek için kullanılıyor.

## roi_hist: Normalleştirilmiş veriyi saklamak için kullanılacak hedef dizi. Yani, bu argüman, işlevin sonucunu saklamak için kullanılır ve genellikle aynı diziye işlem yapılması amaçlanır.

## 0: Normalleştirilen verinin yeni minimum değeri. Yani, veriyi bu değere çekmek için kullanılır. Bu durumda, minimum değer 0 olarak ayarlanıyor.

## 255: Normalleştirilen verinin yeni maksimum değeri. Yani, veriyi bu değere çekmek için kullanılır. Bu durumda, maksimum değer 255 olarak ayarlanıyor.

## cv2.NORM_MINMAX: Normalleştirme yöntemi. Bu argüman, normalleştirme yöntemini belirtir. cv2.NORM_MINMAX, veriyi belirtilen minimum ve maksimum değerlere yeniden ölçekler. Bu, verinin 0 ile 255 arasında bir değere sahip olmasını sağlar, böylece daha sonra işlem yaparken renk skalasının uygun olduğundan emin olunur.


                            TERM CRİT İÇİN OLAN KISIM 

term_crit=(cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT,5,1)

## cv2.TERM_CRITERIA_COUNT: Bu sabit, belirli bir iterasyon sayısına ulaşıldığında takip işleminin durdurulması gerektiğini belirtir. Bu, takip işleminin maksimum iterasyon sayısını kontrol eder.

## term_crit değişkeni bu iki kriteri bir araya getirir ve takip işlemi sırasında bu kriterlerin her ikisinin de sağlanması gerektiğini belirtir. term_crit değişkeninin içeriği aşağıdaki gibidir:



## cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT: Bu iki kriteri bir araya getiren "veya" işlemidir, yani her ikisinin de sağlanması gerektiğini belirtir.
## 5: Maksimum iterasyon sayısıdır. Takip işlemi en fazla 5 iterasyon sürecektir.
## 1: Hata eşiği (epsilon). Eğer takip işlemi sırasında hata eşiği 1'in altına düşerse, işlem durdurulacaktır.
## Bu term_crit değişkeni, nesne takip işlemi sırasında maksimum iterasyon sayısını ve hata eşiğini kontrol etmek için kullanılır. Bu değerler, belirli bir uygulamanın gereksinimlerine ve koşullarına bağlı olarak ayarlanabilir.
